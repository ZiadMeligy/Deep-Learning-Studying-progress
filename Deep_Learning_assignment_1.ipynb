# Generating a Dataset


```python
from random import Random
from matplotlib.pyplot import figure
SEED =5
random_gen = Random(x = SEED)
def generate_pnts(N=1000):
    lst_x, lst_y =[],[]
    for _ in range(N):
        lst_x.append(random_gen.uniform(a=0,b=1))
    for _ in range(N): 
        lst_y.append(random_gen.uniform(a=0,b=1))
    return lst_x,lst_y

data_x, data_y = generate_pnts()
```

<p>Now that the Dataset is generated lets just visualize it for extra points</p>

```python

fig = figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1)
ax.scatter(x=data_x, y=data_y, marker=".");
ax.set_xlim(left=-1,right=2)
ax.set_ylim(top=-1,bottom=2)
fig.show()
```

A perfect square right ? <br>
Now lets make a loss function

```python
from math import sqrt
def loss(x_p,y_p):
    return ((1/len(data_x))*sum(
        [sqrt((x_i-x_p)**2 + (y_i - y_p)**2)
            for x_i,y_i in zip(data_x,data_y)]     
    ))
```

# Numeric Conceptualization of derivatives
Now lets make the gradient descend and optimize our model using numeric conceptualization of derivatives

```python
def find_lowest_loss_arr(x_p,y_p,EPOCHS,H,DELTA):
    epoch_losses=[]
    all_x=[]
    all_y=[]
    for _ in range(EPOCHS):
      all_x.append(x_p)
      all_y.append(y_p)
      epoch_losses.append(loss(x_p,y_p))
      dloss_dx = (loss(x_p+H,y_p)-loss(x_p,y_p))/H
      dloss_dy = (loss(x_p,y_p+H)-loss(x_p,y_p))/H
      x_p -= DELTA * dloss_dx
      y_p -= DELTA * dloss_dy
    return x_p,y_p,epoch_losses,all_x,all_y
```

So here we are going to do that gradient descend method.<br>
optimize and find the point with the lowest loss. <br>
and graph the value of x_p and y_p co-ordinate along the epochs that has been iterated on (gets updated eachh iteration). <br>

# Changing X_P and Y_P to negative number

```python


x_p,y_p =5,5
EPOCHS=3000
DELTA=0.01
H=0.001


epoch_loses =[]
x_corr_list,y_corr_list = [],[]

x_p,y_p,epoch_loses,x_corr_list,y_corr_list=find_lowest_loss_arr(x_p,y_p,EPOCHS,H,DELTA)

fig_xval= figure(figsize=(10,10))
ax = fig_xval.add_subplot(1,1,1)
ax.plot(x_corr_list,'b',y_corr_list,'r')
ax.set_ylabel('value of X_P and Y_P',fontsize=20)
ax.set_xlabel('No. of Epochs',fontsize=20)
fig_xval.suptitle("Loss graph", fontsize = 30)
fig_xval.show()


```
